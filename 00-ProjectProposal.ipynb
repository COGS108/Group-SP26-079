## Research Question

How are transportation barriers and housing vulnerability (index) associated with forgone medical healthcare in Delaware in 2024?

## Background and Prior Work

Transportation barriers are an important obstacle to receiving care, particularly for individuals with lower incomes or those who are uninsured or underinsured. Studies have shown that individuals without reliable transportation are more likely to miss appointments, delay care, or fail to receive recommended treatments <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). These findings show the importance of access to transportation between patients and hospitals.

These transportation challenges are part of a broader framework known as the Social Determinants of Health (SDOH), which refers to non-medical factors that influence health outcomes<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). According to the World Health Organization, SDOH include conditions such as socioeconomic status, education, neighborhood and geographic location, employment, and broader structural forces such as social policies and systemic inequality <a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5). These factors interact with one another to shape access to healthcare services and overall health outcomes. Within this system, transportation and home environment conditions are especially important because they directly affect whether individuals can physically reach healthcare providers and continuously access care.

Prior research has demonstrated consistent relationships between financial barriers, transportation access, and healthcare utilization. Transportation and cost-related barriers are widely recognized as major obstacles to receiving care, particularly among low-income populations and individuals with limited insurance coverage <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). Studies show that when financial barriers are reduced, patients are more likely to access healthcare services, suggesting that cost and access constraints play a key role in determining whether individuals complete the patient journey <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). In a large study of a hospital-based financial assistance program at Kaiser Permanente, researchers found that providing financial assistance significantly increased healthcare utilization, including outpatient visits, emergency department encounters, and inpatient care <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). The study also found that patients receiving financial assistance had higher rates of detection and treatment of health conditions, indicating that reducing financial burden can lead to earlier and more consistent engagement with healthcare services <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). These findings suggest that structural barriers, including both transportation and financial constraints, play a major role in shaping healthcare access and utilization patterns.

Overall, prior work consistently demonstrates that transportation barriers are a key determinant of whether patients are able to successfully complete the healthcare journey. While medical need determines who requires care, transportation access often determines whether that care is actually received. This project builds on existing research by examining how transportation-related barriers are associated with healthcare utilization patterns, highlighting the role of physical access in shaping health outcomes.

## Hypothesis


Null hypothesis: The transportation barrier (measured by vehicle access, etc) and housing vulnerability (measured by government assists in rent, etc) will not affect the healthcare outcome (measured by delayed healthcare, etc).

Alternative hypothesis: The transportation barrier and housing vulnerability will affect the healthcare outcome. 

Our expectation based on previous work: Populations with impeded transportation and uninsured housing may demonstrate a less frequent and standardized routine of accepting healthcare from hospitals and a worse general health status. High housing vulnerability may contribute more than high transportation barriers to the lower than average healthcare outcome.
